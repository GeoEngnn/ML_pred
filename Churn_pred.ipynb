```python
import pandas as pd

# List of categorical columns to encode
cat_cols = ["Contract", "InternetService"]

# One-hot encode categorical columns
x_encoded = pd.get_dummies(x, columns=cat_cols, drop_first=True)

# Now x_encoded contains numeric columns only
print(x_encoded.head())
```

```python
# Fill any potential NaN values with 0 before converting to integer
x_encoded = x_encoded.fillna(0).astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.2)
```

```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np

st.title("Churn Prediction App")
st.divider()
st.write("Please enter the values and hit the predict button for getting a prediction.")
st.divider()
gender = st.selectbox("Select Gender",["Male","Female"])
tenure = st.number_input("Enter Tenure",min_value=0,max_value=100,value=10)
MonthlyCharges = st.number_input("Enter Monthly Charges",min_value=30,max_value=100)
Contract = st.selectbox("Select Contract Type",["Month-to-month","One year","Two year"])
InternetService = st.selectbox("Select Internet Service",["DSL","Fiber optic","No"])
TechSupport = st.selectbox("Select Tech Support",["Yes","No"])
TotalCharges = st.number_input("Enter Total Charges",min_value=1000,max_value=20000)
st.divider()
predictbutton = st.button("Predict")
if predictbutton:
    gender_selected = 1 if gender == "Female" else 0
    Contract_selected = 0
    InternetService_selected = 0
    TechSupport_selected = 0

    if Contract == "One year":
        Contract_selected = 1
    elif Contract == "Two year":
        Contract_selected = 2

    if InternetService == "Fiber optic":
        InternetService_selected = 1
    elif InternetService == "No":
        InternetService_selected = 2

    if TechSupport == "Yes":
        TechSupport_selected = 1

    # Create input array for prediction
    x_input = np.array([[gender_selected, tenure, MonthlyCharges, Contract_selected, InternetService_selected, TechSupport_selected, TotalCharges]])

    # Scale the input data using the fitted scaler
    x_scaled = scaler.transform(x_input)

    # Predict using the best model
    prediction = best_model.predict(x_scaled)[0]

    predicted = "Churn" if prediction == 1 else "Not Churn"
    st.write(f"Prediction: {predicted}")
```
